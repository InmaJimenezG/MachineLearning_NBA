```python
'''
Para completar el método de extacción de los datos de la API de la NBA, se define una función ('NBA_players_data') 
que permite recorrer todas las páginas que contienen los datos personales y estadísticas de los jugadores.
Esta función tiene un único argumento:

        - season = int
        
Este argumento debe referir al año de la temporada de juego sobre la que queremos obtener los datos.
Y devolverá como resultado una variable dónde se recogen todos los datos personales y estadísticas de 
los jugadores:

        - data_players = list
        
'''

def NBA_players_data(season):
    
    import requests
    import json
    
    data_players = []
    
    page = 0
    response = requests.get(f"https://es.global.nba.com/stats2/league/playerstats.json?conference=All&country=All&individual=All&locale=es&pageIndex={page}&position=All&qualified=false&season={season}&seasonType=2&split=All+Team&statType=points&team=All&total=perGame")
    json_response = response.json()

    data_players += json_response["payload"]["players"]
    
    while json_response["payload"]["players"] != []:
        page = page + 1
        response = requests.get(f"https://es.global.nba.com/stats2/league/playerstats.json?conference=All&country=All&individual=All&locale=es&pageIndex={page}&position=All&qualified=false&season={season}&seasonType=2&split=All+Team&statType=points&team=All&total=perGame")
        json_response = response.json()
        data_players += json_response["payload"]["players"]
        
        
    return data_players
```

```python
'''
Por último, se define una función ('season_data') que permite almacenar los datos ligados al año de la temporada
de juego. Esta función tiene un único argumento:

        - season = int
        
El argumento ha de ser igual que el de la función 'NBA_players_data' porque el argumento introducido será
el que se sustitya dentro de la url de la API.
Devolverá como resultado una variable dónde se recogen los datos relacionados con esa temporada de juego:

        - season_list = list
        
'''

def season_data(season):
    
    import requests
    import json
    
    season_list = []
    
    page = 0
    response = requests.get(f"https://es.global.nba.com/stats2/league/playerstats.json?conference=All&country=All&individual=All&locale=es&pageIndex={page}&position=All&qualified=false&season={season}&seasonType=2&split=All+Team&statType=points&team=All&total=perGame")
    json_response = response.json()

    season_list += [json_response['payload']['season']]
        
        
    return season_list
```

```python
'''
Por otro lado, también se define otra función que tiene como objetivo crear un DataFrame por 
cada temporada de juego (df_data_season(dataseason)'.
Como resultado de la misma, se obtiene un DataFrame:

        - df_season = DataFrame
'''


def df_data_season(dataseason):
    
    import pandas as pd
    
    temporada = [] # Para que contenga el año de la temporada sobre la que se crea el DataFrame
    year = [] # Para que contenga el año de comienzo de la temporada
    
    for i in dataseason:
        temporada.append(i['yearDisplay'])
        year.append(int(i['year']))
   

    df_season = pd.DataFrame({'temporada': temporada,
                             'anio': year})

    
    return df_season
```

```python
'''
Primero, se define una función denominada 'df_total'.
Esta función se crea para concatenar por columnas los 2 DataFrames que se han definido  en el apartado 
anterior - 2. Definición de los DataFrames y preparación de los datos - y sustituir los valores NaN de 
la columna 'temporada' por el valor de la primera celda de la misma columna.

Para ello, está función requiere 2 argumentos:

        - df1 = DataFrame
        *En este argumento se añade el DataFrame resultante de la función 'df_data_players(dataplayers)'
        
        - df2 = DataFrame
        *En este argumento se añade el DataFrame resultante de la función 'df_data_season(dataseason)'

'''

def df_total(df1, df2):
    
    df = pd.concat([df2, df1], axis = 1)
    
    for row in range(len(df['temporada'])):
        df['temporada'] = df['temporada'].fillna(df['temporada'][0])
        df['anio'] = df['anio'].fillna(df['anio'][0])
    
    return df
```

```python
'''
Para acceder a la API y obtener un JSON con toda la información, primero se observa que en la url de
la API hay dos parámetros que iteran:

        - 'page': las estadísticas de los jugadores se dividen en varias páginas, ya que son muchos los jugadores
        que forman parte de la NBA (entre 400 - 500 jugadores).
        
        - 'season': refiere a la temporada de juego sobre la que se está obteniendo las estadísticas 
        de todos jugadores.
        
Si se iteran sobre estos dos parámetros, se podrá obtener toda la información de las temporadas ('season')
que queramos y de todos los jugadores integrantes de la NBA durante esa temporada.

Para poder cambiar en la API estos dos parámetros, se definen las siguientes 2 variables:

        - page = int
        - season = int
        
Por otro lado, en la celda anterior se importó la librería 'requests' para poder acceder a la API. 
Se define otra variable para poder usar esta librería e iterear sobre 'page' y 'season':

        - response = requests.models.Response
        
Una vez la variable 'response' está definida con el método 'get' de la libreria 'requests' ('requests.get'), 
se usa el método '.format()' para iterar sobre las variables 'page' y 'season' sobre la url de la API.

Finalmente, para obtener un JSON con los datos de la API, se define otra variable que albergue la 
variable 'response' sobre la que se usa la función '.json()':

        - json_response = dict
        
Lo que resulta de ello es un diccionario que permite ver todos los objetos que contiene el JSON.
'''

# Extraemos la información de los jugadores que aparecen en la página 0, la primera página de estadísticos
page = 0 

# La temporada añadida en esta prueba será la actual (2022-2023), para extraer los estadísticos de juego actuales
season = 2022

response = requests.get(f"https://es.global.nba.com/stats2/league/playerstats.json?conference=All&country=All&individual=All&locale=es&pageIndex={page}&position=All&qualified=false&season={season}&seasonType=2&split=All+Team&statType=points&team=All&total=perGame")
json_response = response.json()

json_response
```
